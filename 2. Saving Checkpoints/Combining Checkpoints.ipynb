```python
import os
import numpy as np
import pandas as pd
```

### 1. Dataset Combining

```python
df = pd.DataFrame()


for file_name in os.listdir('checkpoints/'):
    
    if(file_name != '.DS_Store'):
        
        file_path = 'checkpoints/' + file_name
        
        df_ = pd.read_csv(file_path)
        
        df = pd.concat([df,df_])
        
df.to_csv('data.csv', index = False)
```

```python
product_tag = []
brand_tag   = []

for i in df['product_link']:
    
    product_tag.append(i.split('/')[0].strip())
    brand_tag.append(i.split('/')[1].strip())
    
df['product_tag'] = product_tag
df['brand_tag']   = brand_tag

df.head()
```
