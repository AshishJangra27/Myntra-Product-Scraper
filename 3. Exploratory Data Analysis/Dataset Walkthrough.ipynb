```python
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

### 1. Checking Null Values

```python
df.isnull().sum()
```

### 2. Droping rows with null values

```python
df = df.dropna()

df.isnull().sum()
```

### 3. Checking Dataset with no null values

```python
df.head()
```

### 4. Preprocessing Rating Count

```python
rating_count = []

for i in df['rating_count']:
    
    if(i[-1] == 'k'):
        rating_count.append(int(float(i[:-1]) * 1000))
      
    else:
        rating_count.append(int(i))
        
df['rating_count'] = rating_count

df.head()
```

### 5. Dealing with Marked Price

```python
marked_price = []

for i in df['marked_price']:
    marked_price.append(int(i))
    
df['marked_price'] = marked_price

df.head()
```

### 6. Dealing with Discounted Price

```python
discounted_price = []

for i in df['discounted_price']:
    discounted_price.append(int(i))
    
df['discounted_price'] = discounted_price

df.head()
```

### 8. Adding Discount Column

```python
discount_amount  = []
discount_percent = []

for i in df.values:
    
    
    discount_amount.append(i[4]-i[5])
    discount_percent.append(int(100 - (i[5]/i[4])*100))
    
df['discount_amount']  = discount_amount
df['discount_percent'] = discount_percent

df.head()
```

### 9. Product with Maximum Discount
##### 9.1 Maximum Discount Price

```python
df[df['discount_amount'] == df['discount_amount'].max()]
```

##### 9.2 Maximum Discount Percentage

```python
df[df['discount_percent'] == df['discount_percent'].max()]
```

### 10.Finding the Cheapest Product on Myntra

##### 10.2 Product based on discounted price

```python
df[df['discounted_price'] == df['discounted_price'].min()]
```
