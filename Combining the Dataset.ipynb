```python
import os
import pandas as pd
from tqdm import tqdm
```

### 1. Combining the Dataset with Individual Product Data

```python
files = []

for i in os.listdir('checkpoints_adv/'): 
    if (i != '.DS_Store'):
        files.append(int(i.split('.')[0]))
        
files = sorted(files)

df = pd.DataFrame()


for i in files:
    file_path = 'checkpoints_adv/' + str(i) + '.csv'
    
    df_ = pd.read_csv(file_path)  
    df = pd.concat((df_,df))

df.head()
```

### 2. Load Main Product Data

```python
df_ = pd.read_csv('data.csv')
```

### 3. Combining the Dataset

```python
titles = ['product_name','brand_name','rating','rating_count','marked_price','discounted_price',
          'sizes','product_link','img_link','product_tag','brand_tag','discount_amount','discount_percent',
          'tags','seller_address','meta_description','mera_keyword']

data = []

for i in tqdm(df.values):
    data.append(list(df_[df_['product_link'] == i[0]].values[0]) + list(i[1:]))
```

### 4. Creating the DataFrame

```python
df = pd.DataFrame(data, columns = ['product_name','brand_name','rating','rating_count','marked_price',
                                   'discounted_price', 'sizes','product_link','img_link','product_tag',
                                   'brand_tag','discount_amount','discount_percent', 'tags'
                                   ,'seller_address','meta_description','meta_keyword'])

df.head()
```

### 5. Data Preprocessing

```python
df['tags'] = df['tags'].fillna("['Not Specified']")


tags = []

for i in df['tags']:
    tags.append(','.join([_.replace("'",'').strip() for _ in i[2:-2].split(',')]))
    
df['tags'] = tags

df.head()
```

### 6. Removing Null Values

```python
df['seller_address']    = df['seller_address'].fillna('Not Mentioned')
df['meta_description']  = df['meta_description'].fillna('Not Mentioned')
df['meta_keyword']      = df['meta_keyword'].fillna('Not Mentioned')
```

### 6. Saving the Dataset into CSV

```python
df.to_csv('products.csv', index = False)
```
